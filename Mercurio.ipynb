# **Funciones para nuestros algoritmos de regresion:**

```python
def transformacion(x_train, y_train, x_test, y_test):
  # Transformamos los registros a arreglos numericos pertenecientes a una matriz.
  # Training
  x_train = x_train.to_numpy()
  y_train = y_train.to_numpy().reshape(y_train.size, 1)
  # Testing
  x_test = x_test.to_numpy()
  y_test = y_test.to_numpy().reshape(y_test.size, 1)
  
  # Agregamos las columnas de unos a las matrices x
  # Training
  x_train = np.vstack((np.ones((x_train.shape[0], )), x_train.T)).T
  # Testing
  x_test = np.vstack((np.ones((x_test.shape[0], )), x_test.T)).T
  return x_train, y_train, x_test, y_test
```

```python
def multiple_linear_regression(x, y, alpha, n_iterations):
  n = y.size
  theta = np.zeros((x.shape[1], 1))
  listaCostos = []
  for i in range(n_iterations):
    y_pred = np.dot(x, theta)
    costo = (1/(2*n))*np.sum(np.square(y_pred - y))
    dTheta = (1/n)*np.dot(x.T, y_pred - y)
    theta = theta - alpha*dTheta
    listaCostos.append(costo)
    # to print the cost for 10 times
    if(i%(n_iterations/10) == 0):
      print("Cost is :", costo)
  return theta, listaCostos
```

```python
def grafica_de_costo(listaCostos, n_iterations):
  rango = np.arange(0, n_iterations)
  plt.plot(listaCostos, rango)
  return plt.show()
```

```python
def calculo_error(x,y,theta):
  y_pred = np.dot(x, theta)
  n = x.shape[0]
  error = round((1/n) * np.sum(np.abs(y_pred-y)), 4) * 100
  accuracy = round(100 - error, 4) 
  print("El error en el algoritmo es del: ", error, "%")
  print("La precision del algoritmo es de: ", accuracy, "%")
  return
```
